```python
import numpy as np
import pandas as pd
from shapely.geometry import Point, Polygon
from shapely.wkt import loads
from matplotlib import pyplot as plt
```

```python
df_Wilderness = pd.read_csv("Ecology Project- Wilderness.csv")
df_Settlements = pd.read_csv("Ecology Project- Human Settlements.csv")
df = pd.read_csv('observations-422992.csv')
```

```python
def categorise_pt(str_pt, lat_long=1):
    """
    This function takes a string of coordinates as input and categorises the location as either within a wilderness area,
    within a human settlement, or neither. The categorisation is based on whether the point lies within predefined polygons
    representing wilderness areas and human settlements. Before

    Parameters:
    str_pt (str):       A string of coordinates in the format 'longitude latitude'.
    lat_long:           If the data is input in the format "latitude longitude" then it stays 1. If it is input in the format "longitude latitude" then it is set to not 1.

    Returns:
    If the point is within documented boundaries:
        A tuple with 2 elements. The first element is either Wilderness or Settlement depending on where the observation is found. The second element is the exact sub location on where the observation was.
    If it is not within documented boundaries:
        Returns None

    None (NoneType):    If the coordinate is out of bounds, it returns None

    Usage:
    >>> categorise_pt('79.5885668 13.7126108')
    Location is within WilderNess at Director's Bungalow

    Note:
    This function uses the global dataframes df_Wilderness and df_Settlements, which should contain the WKT polygons
    and names of the wilderness areas and human settlements, respectively.
    """
    coords = str_pt.split()
    if lat_long == 1:
        point = Point(float(coords[0]), float(coords[1]))
    else:
        point = Point(float(coords[1]), float(coords[0]))

    polygon_wild = None
    polygon_settled = None
    wkt_polygon = None

    for i, row in df_Wilderness.iterrows():
        wkt_polygon = row["WKT"]
        polygon_wild = loads(wkt_polygon)

        if polygon_wild.contains(point):
            return "Wilderness", row['name']

    for i, row in df_Settlements.iterrows():
        wkt_polygon = row["WKT"]
        polygon_settled = loads(wkt_polygon)

        if polygon_settled.contains(point):
            return "Settlement", row['name']

    return None
```

```python
def categorise_row(row):
    result = categorise_pt(f"{row['longitude']} {row['latitude']}")
    if result is None:
        return "NOTA"
    else:
        return result[0]  # return either "Wilderness" or "Settlement"
```

```python
df['class_location'] = df.apply(categorise_row, axis=1)
```

```python
a = pivot_df['Wilderness'].sum()  # Total unique species in wilderness
b = pivot_df['Settlement'].sum()  # Total unique species in settlement
j = len(pivot_df[(pivot_df['Wilderness'] > 0) & (pivot_df['Settlement'] > 0)])  # Shared unique species
jn = min(len(pivot_df[(pivot_df['Wilderness'] > 0)]), len(pivot_df[(pivot_df['Settlement'] > 0)]))

print("Parameter a (Unique species in wilderness):", a)
print("Parameter b (Unique species in settlement):", b)
print("Parameter j (Shared unique species):", j)
print("Parameter jn (sum of the lesser values for only those species in common between both sites):", jn)
```

```python
# Jaccard CJ
JCJ = j/(a+b-j)
print(JCJ)
```

```python
# Sorenson CS
SCS = 2*j/(a+b)
print(SCS)
```

```python
#Sorenson Quantative CS
SQCS = 2*jn/(a+b)
print(SQCS)
```

### **Species Accumulation Curves**

```python
# Load the classified data
df_classified = pd.read_csv('observations-422992_classified.csv')

# Convert the 'observed_on' column to datetime
df_classified['observed_on'] = pd.to_datetime(df_classified['observed_on'], format='%Y-%m-%d')

# Initialize the new dataframe
new_df = pd.DataFrame(columns=["observation_date", "number_of_observations", "unique_species_discovered_wilderness", "unique_species_discovered_settlement"])

# Initialize sets to keep track of unique species discovered
unique_species_wilderness = set()
unique_species_settlement = set()

# Group by 'observed_on' and 'class_location'
grouped = df_classified.groupby(['observed_on', 'class_location'])

data = []  # List to store the data

for (date, location), group in grouped:
    number_of_observations = len(group)
    if location == 'Wilderness':
        new_species = set(group['scientific_name']) - unique_species_wilderness
        unique_species_wilderness.update(new_species)
        unique_species_discovered = len(new_species)
    elif location == 'Settlement':
        new_species = set(group['scientific_name']) - unique_species_settlement
        unique_species_settlement.update(new_species)
        unique_species_discovered = len(new_species)
    else:
        continue  # Skip if location is NOTA

    # Add the data to the list
    data.append({
        "observation_date": date,
        "number_of_observations": number_of_observations,
        "unique_species_discovered_wilderness": unique_species_discovered if location == 'Wilderness' else 0,
        "unique_species_discovered_settlement": unique_species_discovered if location == 'Settlement' else 0
    })

# Create the new dataframe
new_df = pd.DataFrame(data)

# Save the new dataframe
new_df.to_csv('sitewise_unique_species_count_by_date.csv', index=False)

# Calculate the cumulative sums
new_df['cumulative_observations'] = new_df['number_of_observations'].cumsum()
new_df['cumulative_species_wilderness'] = new_df['unique_species_discovered_wilderness'].cumsum()
new_df['cumulative_species_settlement'] = new_df['unique_species_discovered_settlement'].cumsum()
```

```python
import pandas as pd
import numpy as np

wilderness_df = df[df['class_location'] == 'Wilderness']

# Calculate the count of each species
species_counts = wilderness_df['scientific_name'].value_counts()

# Create a DataFrame with species counts
count_df = pd.DataFrame({'scientific_name': species_counts.index, 'count_column': species_counts.values})

# Map count_column values from the original DataFrame
count_df['count_column'] = count_df['scientific_name'].map(wilderness_df['scientific_name'].value_counts())

# Calculate the total count of species
sum_counts = count_df['count_column'].sum()

# Calculate the probability of each species occurrence
count_df['probability'] = count_df['count_column'] / sum_counts

# Calculate the relative frequency and log of relative frequency
count_df['n/N'] = count_df['count_column'] / (count_df['count_column'].shift(-1)+a)
count_df = count_df[:-1]
last_division_result_index = count_df['n/N'].last_valid_index()
count_df.at[last_division_result_index, 'n/N'] /= sum_counts
count_df['ln(n/N)'] = np.log(count_df['n/N'])

# Calculate the information entropy for each species
count_df['n/N*ln(n/N)'] = count_df['ln(n/N)'] * count_df['n/N']

# Calculate the sum of n/N*ln(n/N)
sum_all = count_df['n/N*ln(n/N)'].sum()

# Display the sum of n/N*ln(n/N)
print({'Shannon-Wiener Diversity Index_Wilderness': abs(sum_all)})
# print({'sum_of_all_n/N*ln(n/N) negative of it also_wilderness': -sum_all})
```

```python
species_names = ['Ixora coccinea', 'Grewia', 'Passiflora foetida']

rows = []

# Iterate over each species name
for species_name in species_names:
    # Filter DataFrame for the specific species
    species_df = df[df['scientific_name'] == species_name]

    # Check if the species is found in the dataset
    if not species_df.empty:
        # Sum settlement and wilderness counts
        settlement_count = species_df['settlement_count'].sum()
        wilderness_count = species_df['wilderness_count'].sum()

        # Append the row to the list
        rows.append({'scientific_name': species_name,
                     'settlement_count': settlement_count,
                     'wilderness_count': wilderness_count})
    else:
        print(f"The species {species_name} does not occur in the dataset.")

# Convert the list of dictionaries to a DataFrame
counts_df = pd.DataFrame(rows)

# Plotting
plt.figure(figsize=(10, 6))

# Define the width of each bar
bar_width = 0.35

# Define the positions for settlement and wilderness bars
settlement_positions = range(len(counts_df))
wilderness_positions = [pos + bar_width for pos in settlement_positions]

# Plot bars for Settlement counts
plt.bar(settlement_positions, counts_df['settlement_count'], color='orange', label='Settlement', width=bar_width)

# Plot bars for Wilderness counts
plt.bar(wilderness_positions, counts_df['wilderness_count'], color='blue', label='Wilderness', alpha=0.7, width=bar_width)

plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Species Counts in Settlement and Wilderness')
plt.legend()
plt.xticks([pos + bar_width / 2 for pos in settlement_positions], counts_df['scientific_name'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
